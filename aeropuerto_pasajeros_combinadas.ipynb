```python
# =========================================
# ✈️ Análisis de Datos Aeronáuticos
# Este proyecto tiene como objetivo analizar datos de aeropuertos para identificar patrones en el tráfico aéreo y comprender la estructura del sistema aeroportuario.
```

```python
# =========================================
# 1. IMPORTACIÓN DE LIBRERÍAS
# =========================================
import pandas as pd

# =========================================
# 2. RUTA DE ARCHIVOS
# =========================================
ruta = r"C:\Users\ferna\Downloads\python\data\raw"

# =========================================
# 3. CARGA DE DATOS (CORREGIDO)
# =========================================

# Dataset vuelos (con separador correcto y manejo de errores)
df_vuelos = pd.read_csv(
    ruta + r"\aeropuertos_detalle.csv",
    encoding="latin1",
    sep=";",
    on_bad_lines="skip"
)

df_ministerio = pd.read_csv(
    ruta + r"\2020_informe_ministerio.csv",
    encoding="latin1",
    sep=";",
    on_bad_lines="skip",
    low_memory=False
)

print(df_ministerio.info())
print(df_vuelos.info())
```

```python
print(df_vuelos.info())
```

```python
# 🧹 Limpieza de datos
#Se realiza el tratamiento de valores nulos, normalización de variables y corrección de inconsistencias.
```

```python
# Limpiar nombres de columnas
df_vuelos.columns = df_vuelos.columns.str.replace("ï»¿", "")
df_ministerio.columns = df_ministerio.columns.str.replace("ï»¿", "")

# Pasar todo a minúsculas
df_vuelos.columns = df_vuelos.columns.str.lower()
df_ministerio.columns = df_ministerio.columns.str.lower()

# Eliminar espacios extra
df_vuelos.columns = df_vuelos.columns.str.strip()
df_ministerio.columns = df_ministerio.columns.str.strip()

# Renombrar columnas problemáticas manualmente
df_ministerio = df_ministerio.rename(columns={
    "clasificaciã³n vuelo": "clasificacion vuelo"
})

# Ver resultado
print(df_vuelos.columns)
print(df_ministerio.columns)
```

```python
## 🔄 Transformación de datos
# Se seleccionan variables relevantes y se preparan los datasets para el análisis.
```

```python
#Se analizan distribuciones, relaciones entre variables y patrones relevantes en los datos.

# ===============================
# SELECCIÓN DATASET MINISTERIO
# ===============================

df_ministerio = df_ministerio[
    [
        "fecha",
        "hora utc",
        "clasificacion vuelo",
        "tipo de movimiento",
        "aeropuerto",
        "origen / destino",
        "aerolinea nombre",
        "pasajeros",
        "calidad dato"
    ]
]


# ===============================
# SELECCIÓN DATASET VUELOS
# ===============================

df_vuelos = df_vuelos[
    [
        "local",
        "oaci",
        "latitud",
        "longitud",
        "provincia",
        "tipo",
        "trafico"
    ]
]


# Verificar
print(df_ministerio.head())
print(df_vuelos.head())
```

```python
## ⏱️ Ingeniería de variables
#Se generan nuevas variables como hora y franjas horarias para facilitar el análisis.
```

```python
df_ministerio["fecha"] = pd.to_datetime(df_ministerio["fecha"], errors="coerce")
print(df_ministerio["fecha"].head())
print(df_ministerio["fecha"].dtype)
```

```python
def limpiar_texto(texto):
    try:
        return texto.encode("latin1").decode("utf-8")
    except:
        return texto
# Seleccionar columnas de texto
columnas_texto = df_ministerio.select_dtypes(include="object").columns

# Aplicar limpieza
for col in columnas_texto:
    df_ministerio[col] = df_ministerio[col].apply(limpiar_texto)
print(df_ministerio.head())
```

```python
# Seleccionar columnas de texto en df_vuelos
columnas_texto_vuelos = df_vuelos.select_dtypes(include="object").columns

# Aplicar limpieza
for col in columnas_texto_vuelos:
    df_vuelos[col] = df_vuelos[col].apply(limpiar_texto)

# Ver resultado
print(df_vuelos.head())
```

```python
df_ministerio["hora"] = pd.to_datetime(df_ministerio["hora utc"], errors="coerce").dt.hour
print(df_ministerio[["hora utc", "hora"]].head())
```

```python
df_ministerio["hora"] = pd.to_datetime(
    df_ministerio["hora utc"],
    format="%H:%M",
    errors="coerce"
).dt.hour
df_ministerio.head()
```

```python

df_ministerio["hora"].value_counts().sort_index()
```

```python
df_ministerio["hora"].isnull().sum()
```

```python
df_ministerio[
    df_ministerio["hora"].isnull()
][["hora utc"]]
```

```python
df_ministerio["franja_horaria"] = pd.cut(
    df_ministerio["hora"],
    bins=[0, 6, 12, 18, 24],
    labels=["Madrugada", "Mañana", "Tarde", "Noche"],
    right=False
)
```

```python
## 🔗 Integración de datos
#Se combinan los datasets de vuelos y aeropuertos mediante un merge.
```

```python
df_final = pd.merge(
    df_ministerio,
    df_vuelos,
    left_on="aeropuerto",
    right_on="local",
    how="left"
)
```

```python
df_final.isnull().sum()
```

```python
df_final = df_final.drop_duplicates()
```

```python
df_final["provincia"] = df_final["provincia"].fillna("Desconocido")
```

```python
df_final.info()
```

```python
## 📊 Análisis exploratorio
# Se analizan estadísticas descriptivas y distribución de los datos.
```

```python
df_final.describe()
```

```python
df_final = df_final[df_final["pasajeros"] >= 0]
```

```python
print("DF VUELOS:")
print(df_vuelos.head())

print("\nDF MINISTERIO:")
print(df_ministerio.head())
```

```python
df_final["clasificacion vuelo"] = df_final["clasificacion vuelo"].replace({
    "Dom": "Doméstico",
    "Inter": "Internacional"
})
```

```python
## 📈 Análisis de pasajeros por provincia
# Se agrupan los datos para analizar la cantidad de pasajeros por provincia y tipo de vuelo.
```

```python
df_pivot = df_final.pivot_table(
    index="provincia",
    columns="clasificacion vuelo",
    values="pasajeros",
    aggfunc="sum",
    fill_value=0
)

print(df_pivot.head())
```

```python
df_unpivot = df_pivot.reset_index().melt(
    id_vars="provincia",
    var_name="tipo_vuelo",
    value_name="pasajeros"
)

print(df_unpivot.head())
```

```python
print(df_unpivot.info())
print(df_unpivot.describe())
```
